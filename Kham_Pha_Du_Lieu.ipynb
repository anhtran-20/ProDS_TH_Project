### Mỗi cột hiện đang có kiểu dữ liệu gì? Có cột nào có kiểu dữ liệu chưa phù hợp để có thể xử lý tiếp không?

Lấy `dtype` (kiểu dữ liệu của mỗi phần tử) của mỗi cột trong `mvt_df` và lưu kết quả vào series `col_dtypes`; series này có index là tên cột.

**Khám phá các cột có dtype object**

```python
def open_object_dtype(s):
    dtypes=set(s.apply(type))    
    return dtypes

print('- Cột carrier: ')
print(open_object_dtype(df['carrier']))
print('- Cột carrier_name: ')
print(open_object_dtype(df['carrier_name']))
print('- Cột airport: ')
print(open_object_dtype(df['airport']))
print('- Cột airport_name: ')
print(open_object_dtype(df['airport_name']))
```

**Chuyển các cột có kiểu object về kiểu string**

Vì các kiểu dữ liệu chuỗi có độ dài thay đổi, theo mặc định, nó được lưu dưới dạng dtype object. Nếu bạn muốn lưu trữ chúng dưới dạng string, bạn có thể thực hiện như sau.

```python
df['carrier']=df['carrier'].astype('string')
df['carrier_name']=df['carrier_name'].astype('string')
df['airport']=df['airport'].astype('string')
df['airport_name']=df['airport_name'].astype('string')
col_dtypes=df.dtypes
col_dtypes
```
